## Python requests

```python
import requests

url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)

# Print the data (requests can automatically parse JSON for you!)
print(response.json())
```

```python
# HEADERS ONLY
url = "https://jsonplaceholder.typicode.com/users"
response = requests.head(url)

# Print just the headers
for key, value in response.headers.items():
    print(f"{key}: {value}")
```

### Task 2: Base64 Encoding/Decoding with Python
Demonstrating how to convert strings and image files to Base64 and back.

```python
# Import python base64 module
import base64

text = "Hello! This is an example of Base64 encoding."
print("Original Text:", text)

# Encode text
text_bytes = text.encode('utf-8')           # convert text to binary
encoded_text = base64.b64encode(text_bytes) # apply Base64 encoding
print("\nEncoded Text:", encoded_text.decode('utf-8'))
```

```python
# Decode text back to string
decoded_bytes = base64.b64decode(encoded_text)  # decode Base64
decoded_text = decoded_bytes.decode('utf-8')    # convert binary to text
print("Decoded Text:", decoded_text)
```

```python
import base64
import urllib.request
import os

# download a tiny placeholder image so this cell works instantly
image_url = "https://i.pinimg.com/474x/b6/cc/0c/b6cc0c9e32ae561eacd1d74d8a267541.jpg"   # you can replace this
input_img_path = "sample_image.png"
urllib.request.urlretrieve(image_url, input_img_path)
print(f"Downloaded sample image to {input_img_path}")

# Encode image to Base64
with open(input_img_path, "rb") as img_file:
    encoded_image = base64.b64encode(img_file.read())
    # Print just the first 100 characters to avoid flooding the output cell
    print("\nBase64 Encoded Image (first 50 chars):", encoded_image[:50], "...")
    print("\nBase64 Encoded Image (FULL):", encoded_image)

# Decode the Base64 data back into an actual image file
output_img_path = "decoded_sample_image.png"
with open(output_img_path, "wb") as out_img_file:
    out_img_file.write(base64.b64decode(encoded_image))
    print(f"\nDecoded image successfully saved to {output_img_path}")

# Clean up (optional)
# os.remove(input_img_path)
# os.remove(output_img_path)
```
